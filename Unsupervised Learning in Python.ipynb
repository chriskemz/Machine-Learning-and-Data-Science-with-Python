## Unsupervised Learning

Unsupervised learning is a class of machine learning techniques for discovering patterns in data. For intsance, finding the clusters of customers based on their purchase histories, or searching for patterns and correlations among these purchases, and using these patterns to express data in a compressed form. These are examples of unsupervised machine learning technique called __clustering__ and __dimension reduction__ respectively. 

Supervised learning finds patterns for a prediction task. For example, using measurements of tumors to classify them as benign or cancerous. In this case, the pattern discovery is guided or "supervised", so that the patterns are as useful as possible for predicting the label: benign or cancerous. Unsupervised learning, in contrast, is learning without labels. It is a pure pattern dicovery, unguided by a prediction task.

## Evaluating a clustering

A direct approach to evaluate clustering is to check whether the clusters correspond to the samples. We can achieve this using pandas' _crosstab()_ method. The cross-tabulations will provide great insights into which sort of samples are in which cluster. However, in most cases, samples are not labelled. If this information is not given, how can we evaluate the quality of the clustering. Enters __inertia__. Inertia is a metric that measures how internally coherent or tightly packed your clusters are. It calculates the sum of the squared distances of each data point to its closest cluster center (centroid). Thus, it measures how far samples are from the centroids. A good clustering has tight clusters, meaning that samples in each cluster are bunched together, not spread out. Since we want clusters that are not spread out, lower values of inertia are better.

The inertia of a kmeans model is measured automatically when any of the fit methods are called, and is accessed using the _inertia__ attribute. In fact, kmeans aims to place the clusters in a way that minimizes the inertia.

__The Problem with Inertia__ \
Inertia will always decrease as you increase the number of clusters (k). If you pick a small k, say k=1, you have one giant cluster, and the inertia will be very high. If you pick a large k, say k=100, the distance from each sample to its center is very close to or equal to zero, so the inertia will be very close to or equal to zero. This means you can't just pick the k that gives you the lowest inertia, because that would always be the total number of data points, which is useless. So, how do we use inertia to find the best k? Enters the __Elbow method__.

__The Elbow method__ \
The Elbow Method is the standard technique for using inertia to find the optimal number of clusters. The process is simple:
1. Run the K-Means algorithm for a range of different k values (e.g., from 1 to 10).
2. For each k, record the inertia value.
3. Plot the inertia values against the number of clusters (k).

The resulting plot will usually look like an arm. The inertia will drop sharply at first and then start to flatten out. The point where the plot "bends" or "kinks"like an elbowis considered the best value for k. This is the point of diminishing returns, where adding more clusters doesn't significantly reduce the inertia anymore.

## Transforming Features

The variances of features can be a major setback in K-means clustering. The variance of a feature measures the spread of its values. In k-means clustering, the variance of a feature corresponds to its influence on the clustering algorithm. To give every feature a chance, the data needs to be transformed so that features have equal variance. This can be achieved with _StandardScaler()_ from scikit-learn. It tranforms every feature to have a mean of 0 and variance 1. The features are then said to be "standardized".

The APIs of StandardScaler and KMeans are similar, but there's an important difference. StandardScaler tranforms the data, and so it has the transform() method in addition to the fit() method. K-means assigns cluster labels to samples, using the predict() method. K-means also have the fit() method.

Since we need to perform two steps - standardize our data using StandardScaler and cluster the standardized data using k-means - we can conveiently achieve this by combining the two steps using a scikit-learn pipeline. Data then flows from one step to the next, automatically.

StandardScaler is an example of a preprocessing step. _MaxAbsScaler_ and _Normalizer_ are other examples. While StandardScaler() standardizes features by removing the mean and scaling to unit variance, Normalizer() rescales each sample (row) independently of the other, to have unit norm (length)  i.e., the vector representing each data point is scaled to have length 1. Use StandardScaler when you want to standardize features before algorithms that assume normally distributed data or are sensitive to scale, e.g., linear regression, logistic regression, SVM, PCA. Use Normalizer when the direction of the data matters more than the magnitude, such as text classification (TF-IDF vectors), cosine similarity, or when samples have varying scales but you want to compare them as unit vectors.

## Hierarchical Clustering

Hierarchy of groups can be found around us. For example, living things can be organized into small narrow groups, like humans, apes, snakes and lizards; these can then be placed into broader groups like mammals and reptiles, or even broader groups like animals and plants. These groups are contained in one another, and form a hierarchy. Analogously, __hierarchical clustering__ arranges samples into a hierarchy of clusters. Hierarchical clustering is an unsupervised learning method that groups similar data points together, creating a tree-like hierarchy of clusters. Unlike K-Means, you do not need to specify the number of clusters beforehand. The algorithm builds the entire hierarchy for you to explore. 

The result of applying hierarchical clustering to a dataset can be visualized as a tree-like diagram called __dendrogram__. It shows the arrangement of the clusters and the distance at which each merger took place.

__The dendrogram of hierarchical clustering__

The entire process of hierarchical clustering is encoded in the dendogram. At the bottom, every data point is a cluster of its own. The clustering then proceeds from bottom up. Clusters are represented as vertical lines, and a joining of vertical lines a merging of clusters. With 5 data samples, there would be 4 merge operations, and with 6 data samples, there would be 5 merges, and so on. We will use functions from _Scipy_ to perform hierchical clustering. The SciPy _linkage()_ function performs hierarchical clustering on an array of samples. We use _dendrogram()_ to visualize the result.

__How to Read a Dendrogram__:

* The y-axis represents the distance or dissimilarity between clusters. The longer the vertical line, the more dissimilar the clusters were that got merged.
* The x-axis represents the individual data points.
* You read it from the bottom up. Horizontal lines are cluster mergers. The vertical position of the horizontal line indicates the distance at which that merger happened.

__Types of hierarchical clustering__
1. __Agglomerative clustering__: This starts with every data point as its own cluster and progressively merges the closest pairs of clusters until only one large cluster remains.
2. __Divisive clustering__: This is the opposite. It starts with all data points in one giant cluster and recursively splits them into smaller clusters.

We'll focus on the more common agglomerative approach.

__How hierarchical clustering works__

Hierarchical clustering proceeds in steps.
1. Every data point or sample begins in a separate cluster - so there are as many clusters as there are samples 
2. At each step, the two closest clusters are merged. These decreases the number of clusters step by step
3. The process continues until all data points are in a single cluster

## Cluster labels in hierarchical clustering

While the dendrogram is a fantastic visualization, you often need to get the actual cluster labels for each data point to do further analysis. The cluster labels can be extracted from intermediate stages of hierarchical clustering. An intermediate stage in the hierarchical clustering is specified by choosing a height on the dendrogram. The y-axis on the dendrogram (which is the height) encodes the distance between merging clusters. The height that specifies an intermediate clustering corresponds to a distance. This specifies that the hierarchical cluatering should stop merging when all clusters are at least this far apart. 

The distance between two clusters are measured using a __linkage method__. A "complete" linkage indicates that the distance between two clusters is the maximum of the distances between their samples. This was specified via the _method_ parameter of the linkage() function. There are many other linkage methods, and different linkage methods give different hierachical clusterings. For example, in "single" linkage, the distance between clusters is the distance between the closest points of the clusters.

The cluster label for any intermediate stage of the hierarchical clustering can be extracted using the _fcluster()_ function. It returns a Numpy array of cluster labels. Note that Scipy cluster labels starts at 1, not at 0 like in scikit-learn

## Dimensionality Reduction

Dimensionality reduction, or dimension reduction is the process of reducing the number of features (or "dimensions") in a dataset while trying to keep as much of the important information as possible. It finds patterns in data, and uses these patterns to re-express it in a compressed form. Imagine you have a dataset with 50 columns. This can be hard to work with and visualize. Dimensionality reduction aims to simplify it, perhaps down to just 2 or 3 key features that still capture the essence of the original data. This makes subsequent computation with the data much more efficient. The most important function of dimension reduction is to remove noisy features that cause big problems for supervised learning tasks like regression and classification. In many real-world applications, it's dimension reduction that makes prediction possible

__Principal Component Analysis__ \
Principal component analysis is a fundamental dimension reduction technique used to transform a dataset into a new coordinate system, highlighting the directions of greatest variance. Its goal is to create a new, smaller set of features, called __principal components__, that are completely uncorrelated with each other and are ordered by how much information (or "variance") they capture from the original data. 

PCA performs dimension reduction in two steps. The first one called __decorrelation__ will be the focus of this lesson.

__Decorrelation__ \
Decorrelation is the foundation upon which PCA is built. In most real-world datasets, your features are often correlated. For example, in a real estate dataset for Port Harcourt, the size_of_house_sqm and the number_of_bedrooms are likely to be highly correlated. A bigger house usually has more bedrooms. This means the two features provide redundant information. PCA's first step is to transform the data in such a way that these new, transformed features (the principal components) are no longer correlated with each other.

__How decorrelation works__ \
PCA rotates the samples so that they are aligned with the coordinate systems.
1. It identifies the direction in your data where the points are most spread out (the direction of maximum variance). This direction becomes the __First Principal Component__ (PC1).
2. Next, it finds the direction of the next highest variance, with one crucial rule: it must be perpendicular (orthogonal) to PC1. This becomes the __Second Principal Component__ (PC2).
3. This continues for all dimensions. Because all the new axes are perpendicular to each other, the resulting components are, by definition, uncorrelated.

__PCA implementation using scikit-learn__ \
scikit-learn has an implementation of PCA, and it has fit and transform methods just like StandardScaler and KMeans. The fit() method learns how to shift and rotate the samples, but does not actually change them. The transform() method applies the transformation that fit() learned. In particular, the transform() method can be applied to new, unseen samples. We import PCA class from scikit-learn decomposition module. We then create an instance of PCA and fit it to the samples. Then we use the fit PCA object to transform the samples. This will return a new array of transformed samples. This new array has the same number of rows and columns as the original sample array. In particular, there is one row for each transformed sample. The columns of the new arrays are the "PCA features", just as the original features corresponded to the columns of the original array. These resulting PCA features are not linearly correlated. Linear correlation can be measured with the __Pearson correlation__. It takes values between -1 and 1

__Principal components__ \
Principal components are the direction in which the samples vary the most. It the principle components that PCA aligns with the coordinate axes. After a PCA model has been fit, the principal components are available as _component__ attributes. This is a NumPy array for each principal component.

## Intrinsic dimension

The intrinsic dimension of a dataset is the number of features required to approximate it. The intrinsic dimension informs dimension reduction, because it tells us how much a dataset can be compressed. You might have a dataset with 50 columns (an ambient dimension of 50), but if the data points are all constrained to lie on a simple line or a flat sheet within that 50-dimensional space, its intrinsic dimension would be just 1 or 2. 

PCA identifies intrinsic dimension when samples have any number of features. The intrinsic dimension can be identified by counting the PCA features that have high variance. Thus, intrinsic dimension is the number of PCA features that have significant variance.

```python
# finding the direction where data vary the most
# Make a scatter plot of the untransformed points
plt.scatter(grains[:,0], grains[:,1])

# Create a PCA instance: model
model = PCA()

# Fit model to points
model.fit(grains)

# Get the mean of the grain samples: mean
mean = model.mean_

# Get the first principal component: first_pc
first_pc = model.components_[0,:]

# Plot first_pc as an arrow, starting at mean
plt.arrow(mean[0], mean[1], first_pc[0], first_pc[1], color='red', width=0.01)

# Keep axes on same scale
plt.axis('equal')
plt.show()
```

```python
# Perform the necessary imports
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt

# Create scaler: scaler
scaler = StandardScaler()

# Create a PCA instance: pca
pca = PCA()

# Create pipeline: pipeline
pipeline = make_pipeline(scaler, pca)

# Fit the pipeline to 'samples'
pipeline.fit(samples)

# Plot the explained variances
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.xticks(features)
plt.show()
```

From the output above, it looks like PCA features 0 and 1 have significant variance. We can conclude that the dataset has an intrinsic dimension of 2.

## Dimension reduction with PCA

Dimension reduction represents the same data, using less features. It is an important part of building machine learning pipelines. Recall that PCA features are in decreasing order of variance. It performs dimension reduction by discarding the PCA features with lower variance, which it assumes to be noise, and retaining the higher variance PCA features, which it assumes to be informative. To use PCA for dimension reduction, you need to specify how many PCA features to keep, for example, specifying n_components=2 when creating a PCA model tells it to keep only the first two PCA features. A good choice is the intrinsic dimension of the dataset,if you know it.

There are cases where the PCA assumptions do not hold. In these cases, an alternative implementation of PCA needs to be used. __Word frequency arrays__ are a great example. Here, each row corresponds to a document and each column corresponds to a word from a fixed vocabulary. The entries in word frequency arrays measure how often each word appear in each document. Only some of the words from the vocabulary appear in any one document, so most entries are zero. Arrays like this are said to be "sparse", and often represented using a special type of array called a __csr_matrix__. Csr_matrices saves space by remembering only the non-zero entries of the array. Scikit-learn's PCA does not support csr_matrices, and we will need to use _TruncatedSVD_ instead. This performs the same transformation as PCA, but only accepts csr_matrix as input. We can also use the TfidfVectorizer from sklearn. It transforms a list of documents into a word frequency array, which it outputs as a csr_matrix. It has fit() and transform() methods like other sklearn objects.

```python
# Import PCA
from sklearn.decomposition import PCA

# Create a PCA model with 2 components: pca
pca = PCA(n_components=2)

# Fit the PCA instance to the scaled samples
pca.fit(scaled_samples)

# Transform the scaled samples: pca_features
pca_features = pca.transform(scaled_samples)

# Print the shape of pca_features
print(pca_features.shape)
```

We've successfully reduced the dimensionality from 6 to 2
